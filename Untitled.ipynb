```python
#scaling the images to 0-1

def scale_images(data):
    image = data['image']
    return image/255
```

```python
#steps in building a data pipeline in tensorflow includes
#map->cache->shuffle->batch->prefetch [mcsbp]
```

```python
ds = tfds.load('fashion_mnist', split='train')
ds = ds.map(scale_images)
ds = ds.cache()
ds = ds.shuffle(60000)
ds = ds.batch(128)
ds = ds.prefetch(64)
```

```python
ds.as_numpy_iterator().next().shape
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Conv2D, Dense, Flatten, Reshape, LeakyReLU, Dropout, UpSampling2D
```

```python
#building the generator
```

```python
def build_generator():
    model = Sequential()

    #takes in random values and reshapes it to 7x7x128[128 layers]
    #which later we need to convert into a 28x28x1 image
    model.add(Dense(7*7*128, input_dim=128))
    model.add(LeakyReLU(0.2))
    model.add(Reshape((7,7,128)))
    
    #upsampling block 1
    model.add(UpSampling2D())
    model.add(Conv2D(128, 5, padding='same'))
    model.add(LeakyReLU(0.2))


    
    return model
```

```python
test_model = build_generator()
```
